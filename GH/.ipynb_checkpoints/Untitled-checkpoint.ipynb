```python
import librosa
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# audio_path 설정
audio_path = 'test_voice.wav'

# sample late 설정
sr = 44100

# 음성 파일이 정상적으로 설정되었는지 확인
ipd.Audio(audio_path)
```

```python
# auido_path의 음성파일을 librosa 에 load
y, sr = librosa.load(audio_path, sr = sr)

# 각 parameter 값 체크
# sr : sample rate
print('sr:', sr)
# audio의 크기 (sr * time)
print('audio shape:', y.shape)
# audio의 길이
print('length:', y.shape[0]/float(sr), 'secs')
```

```python
# STFT (Short-Time Fourier Transform)
# 정해진 시간 간격 단위(window)로 쪼개 Fourier변환

# n_fft : window 크기
n_fft = 1600

# win_length : 
win_length = 1600

# hop_length : window 간 겹치는 정도
# 일반적으로 n_fft의 1/4
hop_length = 400

stft = librosa.stft(y, n_fft = n_fft, hop_length = hop_length)
spectrogram = np.abs(stft)
print("Spectrogram :\n", spectrogram)

# figure
plt.figure(figsize = (20, 5))
S_dB = librosa.power_to_db(spectrogram, ref=np.max)
librosa.display.specshow(S_dB, sr=sr, hop_length = hop_length, y_axis='linear', x_axis='time')
plt.colorbar(format='%2.0f dB')
plt.show()
```

```python
# MFCC (Mel Frequency Cepstral Coefficient) 추출
mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=20, n_fft=n_fft, hop_length=hop_length)

# log-scale 변환
mfcc = librosa.power_to_db(mfcc, ref=np.max)
```

```python
# pitch 추출
pitch, times = librosa.piptrack(mfcc.shape[0], sr=sr, S=mfcc, fmin=75, fmax=300)
```

```python
import librosa
import matplotlib.pyplot as plt

# Load audio file
filename = "test_voice.wav"
y, sr = librosa.load(filename)

# Compute pitch and magnitude using piptrack
pitch, mag = librosa.piptrack(y=y, sr=sr)

# Compute pitch slope for each section
n = pitch.shape[0]
pitch_slopes = []
for x in range(n):
    pitch_slope = (pitch[n-1, :] - pitch[x, :]) / (n - x)
    pitch_slopes.append(pitch_slope)

# Plot time-pitch_slope graph
plt.figure(figsize=(10, 6))
times = librosa.times_like(pitch)
for i, pitch_slope in enumerate(pitch_slopes):
    plt.plot(times, pitch_slope, color='b', label=f"Section {i}",)
plt.xlabel('Time (s)')
plt.ylabel('Pitch Slope')
plt.title('Time-Pitch Slopes')

plt.tight_layout()
plt.show()

print(pitch_slopes)
```

**3. 형태소 분리**

```python
from kiwipiepy import Kiwi
kiwi = Kiwi()
```

```python
len(temp_lst)
```

```python
#csv 파일 내의 공백 때문에 숫자로 넣음
temp_token = []
for i in range(52):
  result = kiwi.tokenize(temp_lst[i])
 # print(result)
  for x in result:
    temp_token.append(x.form)
```
